#### Decimal Odds

```python
a = []
b = []
c = []

for name in names:
    db = pd.read_csv(f'{name}.csv')
    sns.set(style = 'whitegrid')

    d = db.dropna()
    d.reset_index(drop = True, inplace = True)

    odds = list(d[(np.abs(stats.zscore(d['DecimalOdds'].dropna())) < 3)].index)

    d = d.iloc[odds]

    d = d.sort_values(by = 'Date').reset_index(drop = True)

    x = d.sort_values(by = 'Line').reset_index().Line.values
    y = d.sort_values(by = 'Line').reset_index()['DecimalOdds'].values

    def pred_odds(x, a, b, c):
        return a * np.exp(b * x) + c

    popt, pcov = curve_fit(pred_odds, x, y)

    y_fit = pred_odds(x, *popt)
    
    plt.title(f'{name} Line')
    plt.figure().set_size_inches(5, 2)
    sns.scatterplot(x = x, y = y_fit, data = d)
    
    a.append(popt[0])
    b.append(popt[1])
    c.append(popt[2])
```

```python
def pred_odds(x, a=0.675, b=0.183, c=1.165):
    return a * np.exp(b * x) + c
```

```python
for i in range(len(a)):
    ai = a[i]
    bi = b[i]
    ci = c[i]
    
    db = dbs[i]

    db['Expected Odds'] = pred_odds(db['Line'], ai, bi, ci)
    db['Odds Diff'] = db['DecimalOdds'] - db['Expected Odds']
```
