```python
# An api key is emailed to you when you sign up to a plan
api_key = 'f5c2694d469ba9ee1fab45f36c191327'


# First get a list of in-season sports
sports_response = requests.get('https://api.the-odds-api.com/v3/sports', params={
    'api_key': api_key
})

sports_json = json.loads(sports_response.text)

if not sports_json['success']:
    print(
        'There was a problem with the sports request:',
        sports_json['msg']
    )

else:
    print()
    print(
        'Successfully got {} sports'.format(len(sports_json['data'])),
        'Here\'s all available sports:'
    )
    for i in range(len(sports_json['data'])):
        print(sports_json['data'][i], '\n')



# To get odds for a sepcific sport, use the sport key from the last request
#   or set sport to "upcoming" to see live and upcoming across all sports
sport_key = 'soccer_england_league1'

odds_response = requests.get('https://api.the-odds-api.com/v3/odds', params={
    'api_key': api_key,
    'sport': sport_key,
    'region': 'uk', # uk | us | au
    'mkt': 'h2h' # h2h | spreads | totals
})

odds_json = json.loads(odds_response.text)
if not odds_json['success']:
    print(
        'There was a problem with the odds request:',
        odds_json['msg']
    )

else:
    # odds_json['data'] contains a list of live and 
    #   upcoming events and odds for different bookmakers.
    # Events are ordered by start time (live events are first)
    print()
    print(
        'Successfully got {} events'.format(len(odds_json['data'])),
        'Here\'s the events:'
    )
    for i in range(len(odds_json['data'])):
        print(odds_json['data'][i], '\n')

    # Check your usage
    print()
    print('Remaining requests', odds_response.headers['x-requests-remaining'])
    print('Used requests', odds_response.headers['x-requests-used'])

```

```python
with open('oddstest.json', 'r') as read_file:
    data= json.load(read_file)
```

```python
for dct in data['data']:
    print('')
    print(dct['sport_key'])
    print (dct['sport_nice'])
    print(dct['teams'])
    print(datetime.fromtimestamp(dct['commence_time']))
    lst = [site['odds']['h2h'] for site in dct['sites']]
    side1 = []
    side2 = []
    tie = []
    for book in lst:
        print(book)
        side1.append(book[0])
        side2.append(book[1])
        if len(book)>2:
            tie.append(book[2])
    try:
        
        print('Side 1 Odds', mean(side1))
        print('Side 2 Odds', mean(side2))
        if len(tie)>0:
            print('Tie Odds', mean(tie))
            print('')
    except: 
        print('no odds')
        
```
