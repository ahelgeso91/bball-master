```python
def get_nba_lines():
    
    spread_lines('basketball_nba')
    with open('spreadLines.json','r') as read_file:
        data = json.load(read_file)
    
    team1 = []
    team2 = []
    location = []
    date = []
    odds1 = []
    odds2 = []
    line1 = []
    line2 = []
    at = []

    for game in data['data']:
        date.append(datetime.fromtimestamp(game['commence_time']).strftime('%m/%d/%y'))
        team1.append(game['teams'][0])
        team2.append(game['teams'][1])
        location.append(game['home_team'])

        line1_temp = []
        line2_temp = []

        for site in game['sites']:
            line1_temp.append(float(site['odds']['spreads']['points'][0]))
            line2_temp.append(float(site['odds']['spreads']['points'][1]))

        line1.append(mean(line1_temp))
        line2.append(mean(line2_temp))

    for i in range(len(location)):
        if team1[i]==location[i]:
            at.append(' ')
        else:
            at.append('@')
    nba_lines = pd.DataFrame({'Date':date,'Team1':team1,'At':at, 'Team2':team2,'Line1':line1,'Line2':line2})
    nba_lines.to_csv('NBA Lines.csv',index = False)
    return nba_lines
```
